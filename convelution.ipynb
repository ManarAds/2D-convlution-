```python
import numpy as np
import cv2
import matplotlib.pyplot as plt 


x=np.array([[1,0,0], [0,0,1], [1,0,0]])
print(x)

plt.imshow(x , cmap='binary')
```

```python
filter=np.array([[1,1], [1,1]])
filter
```

```python
from scipy.signal import convolve2d
convel=convolve2d(x,filter)
convel
```

```python
plt.figure(1)
plt.subplot(1,3,1) #1*3 first plot 
plt.imshow(x , cmap='binary')


plt.subplot(1,3,2)
plt.imshow(filter, cmap='binary')

plt.subplot(1,3,3)
plt.imshow(convel, cmap='binary')
```
